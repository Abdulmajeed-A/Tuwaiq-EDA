# Mental Health in the Workplace Data Exploration

**Data:** This dataset, typically known as the OSMI Mental Health in Tech Survey, explores the intersection of mental health and the workplace within the technology industry. It aims to quantify how mental health issues are perceived, treated, and supported in professional environments.

`Link:` https://www.kaggle.com/datasets/osmi/mental-health-in-tech-survey


## Dataset Columns Description:

This dataset contains survey responses related to mental health in the workplace.

- **Timestamp** – Time when the survey response was submitted.  
- **Age** – Age of the respondent.  
- **Gender** – Gender identity reported by the respondent.  
- **Country** – Country where the respondent resides.  
- **state** – U.S. state or territory of residence (only applicable for respondents in the United States).  
- **self_employed** – Indicates whether the respondent is self-employed.  
- **family_history** – Whether the respondent has a family history of mental illness.  
- **treatment** – Whether the respondent has sought treatment for a mental health condition.  
- **work_interfere** – Indicates if a mental health condition interferes with the respondent’s work.  
- **no_employees** – Number of employees in the respondent’s company or organization.  
- **remote_work** – Whether the respondent works remotely at least 50% of the time.  
- **tech_company** – Indicates if the employer is primarily a technology company.  
- **benefits** – Whether the employer provides mental health benefits.  
- **care_options** – Whether the respondent knows the mental health care options provided by their employer.  
- **wellness_program** – Whether the employer includes mental health discussions in employee wellness programs.  
- **seek_help** – Whether the employer provides resources for learning about mental health and seeking help.  
- **anonymity** – Whether anonymity is protected when using mental health treatment resources.  
- **leave** – Ease of taking medical leave for a mental health condition.  
- **mental_health_consequence** – Whether discussing mental health with an employer may have negative consequences.  
- **phys_health_consequence** – Whether discussing physical health with an employer may have negative consequences.  
- **coworkers** – Willingness to discuss mental health issues with coworkers.  
- **supervisor** – Willingness to discuss mental health issues with a direct supervisor.  
- **mental_health_interview** – Whether the respondent would bring up mental health in a job interview.  
- **phys_health_interview** – Whether the respondent would bring up physical health in a job interview.  
- **mental_vs_physical** – Whether the employer treats mental health as seriously as physical health.  
- **obs_consequence** – Whether the respondent has observed negative consequences for coworkers with mental health conditions.  
- **comments** – Additional notes or comments provided by the respondent.

## Data Cleaning
Based on our initial inspection, we will apply the following cleaning steps:
* Handle missing values (e.g., dropping or filling 'state' and 'work_interfere').
* Remove duplicates.
* Clean and standardize the 'Gender' column (which is known to have messy text entries).
* Convert data types where necessary.
* Save the cleaned data to a new CSV file.

```python
# 1. Remove duplicates
df = df.drop_duplicates()
```

```python
# 2. Handle Missing Values

# 'comments' has too many missing values, so we will drop the column entirely
df = df.drop(columns=['comments'])

# 'state' is only relevant for US, we'll fill NA with 'Unknown'
df['state'] = df['state'].fillna('Unknown')

# 'work_interfere' is a useful metric, we'll fill missing values with 'Not sure'
df['work_interfere'] = df['work_interfere'].fillna('Not sure')

# 'self_employed' has a few missing, fill with the mode ('No')
df['self_employed'] = df['self_employed'].fillna('No')

# Make sure no more missing values
print("\n--- Missing Values Count ---")
print(df.isnull().sum())
```

```python
# 3. Clean the 'Gender' column (Standardizing messy text inputs)

df['Gender'] = df['Gender'].str.lower().str.strip()
male_terms = ['male', 'm', 'make', 'man']
female_terms = ['female', 'f', 'woman']

def clean_gender(x):
    if x in male_terms:
        return 'Male'
    elif x in female_terms:
        return 'Female'
    else:
        return 'Other'

df['Gender'] = df['Gender'].apply(clean_gender)
df = df[df["Gender"] != "Other"]

# Make sure did it right
print(df["Gender"].unique())
```

```python
# 4. Filter impossible Age values (e.g., negative ages or ages over 100)
df = df[(df['Age'] >= 18) & (df['Age'] <= 100)]

# Make sure
df.describe()
```

## Answer research questions (2)

### 1- How do workplace factors (such as benefits, leave policies and anonymity) influence whether employees seek mental health treatment?

Workplace Factors vs. Seeking Treatment


```python
# Define consistent order and y-axis range
order = ['Yes', "Don't know", 'No']
y_limit = (0, 450)

# 1. Impact of Benefits on Treatment
plt.figure(figsize=(10, 5))
sns.countplot(
    data=df,
    x='benefits',
    hue='treatment',
    palette='Set2',
    order=order
)
plt.title('Impact of Mental Health Benefits on Seeking Treatment')
plt.xlabel('Does your employer provide mental health benefits?')
plt.ylabel('Count')
plt.ylim(y_limit)
plt.legend(title='Sought Treatment', loc='upper right')
plt.tight_layout()
plt.show()


# 2. Impact of Anonymity on Treatment
plt.figure(figsize=(10, 5))
sns.countplot(
    data=df,
    x='anonymity',
    hue='treatment',
    palette='Set1',
    order=order
)
plt.title('Impact of Anonymity on Seeking Treatment')
plt.xlabel('Is anonymity protected?')
plt.ylabel('Count')
plt.ylim(y_limit)
plt.legend(title='Sought Treatment', loc='upper right')
plt.tight_layout()
plt.show()
```

**Insight:** Employees who know their company provides mental health benefits, guarantees anonymity, and offers easy medical leave are significantly more likely to seek treatment. Interestingly, you will likely find that a large portion of employees answer "Not sure" to these questions, and this group tends to have lower treatment rates. This highlights that communicating benefits is just as important as having them.

### 2- Does discussing mental health at work carry greater perceived negative consequences than discussing physical health?

Mental vs. Physical Health Stigma


```python
# Ensure consistent category order
order = ['Yes', 'No', 'Maybe']

# Create figure with two side-by-side subplots
fig, axes = plt.subplots(1, 2, figsize=(14, 5))

# Mental Health Consequences
sns.countplot(
    data=df,
    x='mental_health_consequence',
    hue='mental_health_consequence',
    palette='Reds',
    order=order,
    legend=False,
    ax=axes[0]
)
axes[0].set_title('Perceived Negative Consequence: Mental Health')
axes[0].set_xlabel('Would discussing it have negative consequences?')
axes[0].set_ylabel('Count')
axes[0].set_ylim(0, 900)  # Set y-axis range

# Physical Health Consequences
sns.countplot(
    data=df,
    x='phys_health_consequence',
    hue='phys_health_consequence',
    palette='Blues',
    order=order,
    legend=False,
    ax=axes[1]
)
axes[1].set_title('Perceived Negative Consequence: Physical Health')
axes[1].set_xlabel('Would discussing it have negative consequences?')
axes[1].set_ylabel('Count')
axes[1].set_ylim(0, 900)  # Set y-axis range

# Tidy layout and show
plt.tight_layout()
plt.show()
```

**Insight:** Yes, absolutely. This is usually the most striking finding in this dataset. When you compare mental_health_consequence to phys_health_consequence, the vast majority of respondents feel that discussing physical health has "No" negative consequences. However, when asked about mental health, the responses heavily shift toward "Maybe" and "Yes."
